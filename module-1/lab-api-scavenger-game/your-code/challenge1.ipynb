```python
import requests
import json
import os
from dotenv import load_dotenv
```

```python
#Hacemos la funcion para obtener la informacion a buscar de la p√°g
def get_gh(endpoint):
  
    baseUrl = "https://api.github.com"
    url = f"{baseUrl}{endpoint}"
    print(f"Request data to {url}")

    res = requests.get(url)
    data = res.json()
    return data
```
